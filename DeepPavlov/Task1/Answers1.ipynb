**1. You roll a biased coin with p(head)=0.65 seven times. What is the probability of getting more than two heads?**

Потребуется найти суммарную вероятность выпадения 3, 4, 5, 6 и 7 монет одновременно.

$P_{sum} = P_3 + P_4 + P_5 + P_6 + P_7$

Для нахождения вероятности слагаемых воспользуемся формулой Бернулли:

$P_{k,n}=C_n^k\cdot p^k \cdot q^{n-k}$, где $q = 1 - p$

$P_3 = \frac{n!}{k!(n-k)!} * p_{head}^k *(1-p_{head})^{n-k} = \frac{7!}{3!(7-3)} * 0.65^3 *(1-0.65)^{7-3} = 0,144238199$

$P_4 = ... = 0,267870941$, 

$P_5 = ... = 0,298484763$, 

$P_6 = ... = 0,184776282$,

в случае с $P_7$ первый множитель в формуле Бернулли приравниваем к единице (иначе в знаменателе получим деление на 0) и используем эту формулу как в предудущих случаях

$P_7 = 1 * p_{head}^k *(1-p_{head})^{n-k} = 0,049022279$.

Итого ответ:

$P_{sum} = P_3 + P_4 + P_5 + P_6 + P_7 = 0,944392464$.
